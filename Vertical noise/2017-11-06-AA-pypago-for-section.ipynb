```python
gridname='/media/extra/DATA/NACHOS12.L75/NACHOS12.L75-I/NACHOS12.L75-MAA13_mesh_mask.nc'

lont=np.squeeze(pypago.readnc(gridname,'glamt'));
latt=np.squeeze(pypago.readnc(gridname,'gphit'));
zt=np.squeeze(pypago.readnc(gridname,'gdept_1d'));
bathy_temp=np.squeeze(pypago.readnc(gridname,'mbathy')).astype(np.int);
bathy_temp[bathy_temp<0]=0;
bathy=zt[bathy_temp]
mask_temp=np.squeeze(pypago.readnc(gridname,'tmask'));
mask=np.squeeze(mask_temp[0,:,:]);
```

```python
[nj,ni]=mask.shape
latlim=[1,mask.shape[0]-2]
lonlim=[0,mask.shape[1]-1]


```

```python
MODEL_grid={'model':'NEMO',
            'lonw':lonlim[0],
            'lone':lonlim[-1],
            'lats':latlim[0],
            'latn':latlim[-1]}       
MODEL_grid=pypago.pago_obj(MODEL_grid) # we convert the dict. to pago object
```

```python
dxt=np.squeeze(pypago.readnc(gridname,'e1t')); # at center 
dyt=np.squeeze(pypago.readnc(gridname,'e2t')); # at center
dyte=np.squeeze(pypago.readnc(gridname,'e2u')); # at eastern face of cell
dxtn=np.squeeze(pypago.readnc(gridname,'e1v')); # at northern face of cell
dzt=np.squeeze(pypago.readnc(gridname,'e3t_0')); # at center
dze=np.squeeze(pypago.readnc(gridname,'e3u_0')); 
dzn=np.squeeze(pypago.readnc(gridname,'e3v_0'));
dzn=np.ma.array(dzn,mask=dzn!=dzn)            
dze=np.ma.array(dze,mask=dze!=dze)      
dze[dze.mask==True]=np.nan
dzn[dzn.mask==True]=np.nan
dzw=np.zeros(dze.shape)
dzw[:,:,0]=dze[:,:,-1]
dzw[:,:,1:]=dze[:,:,0:-1]
```

```python
MODEL_grid.lont=lont[MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]
MODEL_grid.latt=latt[MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]
MODEL_grid.bathy=bathy[MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]
MODEL_grid.mask=mask[MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]

```

```python
dxt=dxt[MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]
dxtn=dxtn[MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]
dyt=dyt[MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]
```

```python
dytw=np.zeros(dyte[MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1].shape)   
dytw[:,0]=dyte[MODEL_grid.lats:MODEL_grid.latn+1,-1]
dytw[:,1:]=dyte[MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone]
```

```python
dzt=dzt[:,MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]
dzw=dzw[:,MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]
dzn=dzn[:,MODEL_grid.lats:MODEL_grid.latn+1,MODEL_grid.lonw:MODEL_grid.lone+1]
```

```python
[nz,ni,nj]=dzt.shape
```

```python
areaW=dzw*np.tile(dytw,[nz,1,1])
areaN=dzn*np.tile(dxtn,[nz,1,1])

volume=dzt*np.tile(dxt*dyt,[nz,1,1])
volume[volume!=volume]=0
surface=dxt*dyt
```
